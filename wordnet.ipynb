```python
def get_links(word_id='n04194289'): 
    """
        Get images links from a word_id (example : n04194289)
    """
    page = requests.get("http://www.image-net.org/api/text/imagenet.synset.geturls?wnid=" + word_id)#ship synset
    return BeautifulSoup(page.content, 'html.parser')
```
